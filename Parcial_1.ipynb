a) La distancia media entre dos señales periódicas $x_1(t)\in\mathbb{R}$ y $x_2(t)\in\mathbb{R}$, se puede expresar a partir de la potencia media de la diferencia entre ellas:

$$
d^{2}(x_1, x_2) = \bar{P}_{x_1-x_2}=\lim_{T \to \infty} \frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}~dt.
$$

Sea $x_1(t)$ y $x_2(t$) dos señales como se muestra a continuación:

$$
x_1(t) = Ae^{jw0t}
$$

$$x_2(t)= Be^{j5w0t}
$$
Con
$$  w_0 = \frac{2\pi}{T} ; T, A, B \in \mathbb{R}^+ .
$$
Determine la distancia entre las dos señales.

### Solución

Para calcular la distancia entre dos señales periódicas $x_1(t)$ y $x_2(t)$, utilizaremos la siguiente fórmula:
$$
d^2(x_1, x_2) = \lim_{T \to \infty} \frac{1}{T} \int_{0}^{T} |x_1(t) - x_2(t)|^2 \, dt
$$
Dadas las señales:
- $x_1(t) = Ae^{j\omega_0 t}$
- $x_2(t) = Be^{5j\omega_0 t}$

donde $\omega_0 = \frac{2\pi}{T}$, y $T, A, B \in \mathbb{R}^+$.

#### Paso 1: Integramos y usamos la formula de B.C.P

$$
\frac{1}{T} \left[\int_0^T  |x_1(t)|^2 \,dt -2  \int_0^T x_1(t)x_2(t)\,dt + \int_0^T  |x_2(t)|^2 \,dt\right]
$$
#### Nuestra ecuacion quedaria:

$$ P_{x_1}  - \frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right] + P_{x_2}$$


#### Paso 2: Calculamos  $ P_{x_1}$ y $P_{x_2}$

$ P_{x_1}$

$$
P_{x_1} = \frac{1}{T} \int_{0}^{T} |x_1(t)|^2 \, dt
$$

$$
\frac{1}{T} \int_{0}^{T} |Ae^{j\omega_0 t}|^2 \, dt $$

$$ \frac{A^2}{T} \int_{0}^{T} e^{ j\omega_0 t} \cdot e^{-j\omega_0 t} \, dt$$

$$ \frac{A^2}{T} \int_{0}^{T} e^{j\omega_0 t-j\omega_0 t}  \, dt$$

$$\frac{A^2}{T} \int_{0}^{T} e^{0} \, dt $$

$$\frac{A^2}{T} \int_{0}^{T} 1 \, dt$$

$$ \frac{A^2}{T} [t]_{0}^{T}$$

$$\frac{A^2}{T} [T] = A^2$$

$$P_{x_1} = A^2$$




 $ P_{x_2}$

$$
P_{x_2} = \frac{1}{T} \int_{0}^{T} |x_2(t)|^2 \, dt
$$

$$
\frac{1}{T} \int_{0}^{T} |Be^{5j\omega_0 t}|^2 \, dt $$

$$ \frac{B^2}{T} \int_{0}^{T} e^{ 5j\omega_0 t} \cdot e^{-5j\omega_0 t} \, dt$$

$$ \frac{B^2}{T} \int_{0}^{T} e^{5j\omega_0 t-5j\omega_0 t}  \, dt$$

$$\frac{B^2}{T} \int_{0}^{T} e^{0} \, dt $$

$$\frac{B^2}{T} \int_{0}^{T} 1 \, dt$$

$$ \frac{B^2}{T} [t]_{0}^{T}$$

$$\frac{B^2}{T} [T] = B^2$$

$$P_{x_2} = B^2$$


#### Procedemos a encontramos el valor de $\frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right]$

$$
\frac{2}{T} \int_{0}^{T} x_1(t) x_2(t) dt $$

$$\frac{2}{T} \int_{0}^{T} (Ae^{j\omega_0 t})(Be^{j5\omega_0 t}) dt = \frac{2AB}{T} \int e^{6j\omega_0 t} dt
$$

tenemos que $\omega_0 = \frac{2\pi}{T}$ , entonces lo reemplazamos en la ecuacion
$$
\frac{2AB}{T} \int e^{-12j \frac{\pi}{T} t} dt
$$
Se aplica un cambio de variable
$$
u = 6 \frac{2\pi}{T} t
$$
$$
$$
$$
du = \frac{12j\pi }{T} dt
$$
$$
$$
$$
dt = \frac{T}{12j\pi} du
$$
La ecuacion quedaria:
$$
   \frac{2AB}{T} \int e^{u} \frac{T}{12j\pi} du
$$
Resolvemos la ecuacion
$$
\frac{2AB}{T} \frac{T}{12j\pi} \int_0^T e^u du
$$
$$
\frac{AB}{6j\pi} \left[ e^{{12j\pi}}  e^{0}\right]
$$

$$
= \frac{AB}{6j\pi} \left[ e^{{12j\pi}} - 1 \right]
$$


#### Por ultimo, remplazamos los valores que hallamos en la ecuacion original:
$$ P_{x_1-x_2} =  P_{x_1}  - \frac{2}{T} \left[ \int_0^T x_1(t)x_2(t)\,dt \right] + P_{x_2}$$

Asi:

$$
d^{2}{(x_1,x_2)} = A^2-\frac{AB}{6j\pi} \left[ e^{{12j\pi}} - 1 \right] +B^2
$$


#### Conclusión

La distancia media cuadrada entre las señales $x_1(t)$ y $x_2(t)$ es la suma de los cuadrados de las amplitudes $A$ y $B$, lo cual muestra que la distancia no depende de las frecuencias de las señales y tampoco de la variable de tiempo $t$.

```python


from sympy import *

# Definimos los símbolos
t, T, A, B, w0 = symbols('t T A B w0', real=True, positive=True)

# Definimos las señales
x1 = A * exp(I * 2*pi/T* t)
x2 = B * exp(I * 5 * 2*pi/T * t)

# Calculamos la diferencia entre las señales
diff_signal = abs(x1 - x2)

# Calculamos la potencia media de la diferencia
P_mean_diff =  1/T * integrate(diff_signal**2, (t, 0, T))

# Imprimimos el resultado
print('La potencia media de la diferencia entre las señales es:')
P_mean_diff.simplify()
```

b). ¿Cuál es la señal obtenida en tiempo discreto al utilizar un conversor analógico digital con frecuencia de muestreo de  $5kHz$,, aplicado a la señal continua
$$
x(t) = 3\cos (1000\pi t) + 5\sin (2000\pi t) + 10\cos (11000\pi t)
$$

Realizar la simulación del proceso de discretización. En caso de que la discretización no sea apropiada, diseñe e implemente un conversor adecuado para la señal estudiada.


### Solución:

Se reemplaza $t$ en continuo por la discretización $t=nT_s$, con $n\in\mathbb{Z}$:

$x[t = nT_s] = 3\cos[1000\pi n T_s]+ 5\sin[2000\pi n T_s] + 10\cos[11000\pi n T_s],$

teniendo en cuenta que:

$T_s = \frac{1}{F_s} = \frac{1}{5000} [s].$

Ahora:


$x[n] = 3\cos[1000\pi n/5000]+ 5\sin[2000\pi n/5000] + 10\cos[11000\pi n / 5000]$

$x[n] = 3\cos[\pi n/5] + 5\sin[2\pi n/5] + 10 \cos[11\pi n /5]$

$\Omega_3 = 11\pi/5 > \pi,$ frecuencia copia!

Para estimar la frecuencia original:

$\boxed{\hat{\Omega}_3 = \Omega_3 - 2\pi = 11\pi/5 -2\pi = \pi/5}$

Reemplazando en la función discretizada:


$x[n] = 3\cos[\pi n/5] + 5\sin[2\pi n/5] + 10 \cos[\pi n /5]$

Tenemos que:

$\boxed{x[n] = 13\cos[\pi n /5]+ 5 \sin[2\pi n /5]}$

```python

import numpy as np
import matplotlib.pyplot as plt

# Definimos la señal x(t) como una función
def x(t):
    return 3 * np.cos(1000 * np.pi * t) + 5 * np.sin(2000 * np.pi * t) + 10 * np.cos(11000 * np.pi * t)

# Establecemos la frecuencia de muestreo
fs1 = 5000  # 5 kHz para la simulación incorrecta
fs2 = 22000 # 22 kHz para evitar aliasing (el doble de la máxima frecuencia más un margen)

# Generamos un vector de tiempo continuo y discreto
t = np.linspace(0, 0.008, 1000)  # 8 ms de señal para una buena visualización
t1 = np.arange(0, 0.008, 1/fs1)  # Muestreo incorrecto
t2 = np.arange(0, 0.008, 1/fs2)  # Muestreo correcto para evitar aliasing

# Evaluamos la señal en los puntos de muestreo
x_t = x(t)
x_t1 = x(t1)
x_t2 = x(t2)

# Realizamos la gráfica de la señal continua y las señales discretizadas
plt.figure(figsize=(15, 5))

# Señal continua
plt.plot(t, x_t, label='Señal continua x(t)')

# Señales muestreadas
plt.stem(t1, x_t1, 'r', markerfmt='ro', basefmt=" ", linefmt='r--', label='Muestreo a 5 kHz (incorrecto)')
plt.stem(t2, x_t2, 'g', markerfmt='go', basefmt=" ", linefmt='g--', label='Muestreo a 22 kHz (correcto)')

# Mostramos la leyenda
plt.legend()

# Títulos y etiquetas
plt.title('Simulación de muestreo de la señal x(t)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
# Mostrar la gráfica
plt.show()
```

Conclusion:

La línea continua representa la señal original, los puntos rojos representan el muestreo incorrecto a 5 kHz, y los puntos verdes representan el muestreo correcto a 22 kHz, el cual es adecuado para evitar el aliasing.

Podemos observar que con el muestreo incorrecto a 5 kHz, no todos los ciclos de la señal son capturados correctamente, lo que puede llevar a una representación errónea de la señal cuando se reconstruya. En cambio, con el muestreo a 22 kHz, se capturan más puntos de la señal, lo cual permitiría una reconstrucción más precisa de la misma.

c). Implemente una simulación para encontrar la salida del sistema lineal e invariante al tiempo  $\mathcal{H}\{\cdot\}$ , con respuesta al escalón  $ h_\epsilon[n] = {(2, 4, 1, 5, 0, 10)} , n \in\mathbb{Z}$, ante la entrada analógica en corriente
$$ x(t) = 20(\cos(t/3) + \cos(t/4)) [A]$$ Incluya los acondicionamientos necesarios de discretización y cuantización, asumiendo un microprocesador de 4 bits con entrada analóga de 4mA a 20mA.

## Solución:

Revisando la señal $x(t)$, se tiene que:

$\frac{\omega_1}{\omega_2} =\frac{1/3}{1/4} = \frac{4}{3} \in \mathbb{Q}$; por ende $x(t)$ es cuasiperiódica, con período:

$T_1 = \frac{2\pi}{\omega_1} = \frac{2\pi}{1/3} = 6\pi$

$T_2 = \frac{2\pi}{\omega_2} = \frac{2\pi}{1/4} = 8\pi$

$T = T_1 l = T_2 k$

$T = 6\pi l = 8\pi l$

con $k,l\in \mathbb{Z}.$ El $MCM(6,8)=24,$ así:

$\boxed{T = 24\pi [s]}$

Además, se requiere aplicar cero pendiente:

$\hat{x}(t)=m x(t)+ b$, con:

$m = \frac{\hat{x}_{max}(t)-\hat{x}_{min}(t)}{x_{max}(t)-x_{min}(t)}$

$b = \hat{x}_{min}(t)-m x_{min}(t)$

```python
#función cero y pendiente
def my_ceropend(x,xhmax=0.02,xhmin=0.004):
  #cálculo de pendiente y cero
  m = (xhmax-xhmin)/(x.max()-x.min())
  b = xhmin - m*x.min()
  xh = m*x+b
  return xh
```

Se implementan los códigos de discretización y cuantización:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

T1 = 6*np.pi #perido señal 1
T2 = 8*np.pi #perido señal 2
F1 = 1/T1
F2 = 1/T2
T = 24*np.pi #perido cuasiperiódica
Fs = 10*max(F1,F2) #frecuencia muestreo
t = np.arange(0,3*T,1/Fs) #vector de tiempo, se muestran 3 periodos
x = 20*(np.cos(t/3)+np.cos(t/4))
xhmax = 0.02
xhmin = 0.004
xh = my_ceropend(x,xhmax, xhmin).reshape(-1,1)#aplicar cero y pendiente de 0 a 10
#asegurar vector columna

#cuantizar
nbits = 4 #número bits
Ne = 2**nbits #cantidad de estados
res = (xhmax-xhmin)/Ne #resolución cuantización
ve = np.arange(xhmin,xhmax,res).reshape(-1,1)#vector de estados
#asegurar vector columna

xq = np.zeros(len(xh)) #señal cuantizada
for i in range(len(xh)): #recorrer señal cero y pendiente
    xq[i] = ve[np.argmin(cdist(xh[i].reshape(-1,1),ve))] #se cálcula la distancia desde cada punto al vector de estados

#gráficas de señales
plt.title('Señal original')
plt.plot(t,x,'r',label='$x(t)$')
plt.xlabel('t [s]')
plt.ylabel('Amplitud')
plt.show()
plt.plot(t,xh,'b',label='S.Original')
plt.title('Señal cuantizada')
plt.stem(t,xq,'y',label='$x_{q}[n]$')
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.legend()
plt.show()
```

Teniendo la señal discretizada y cuantizada, se procede a encontrar la salida del SLIT aplicando la convolución entre la entrada digital y la respuesta impulso. Dado que tenemos la respuesta escalón, calculamos la respuesta impulso como:

$h[n] = {\rm{diff}}[h_\epsilon[n]] = h_\epsilon[n]-h_\epsilon[n-1]$

```python
he = np.array([2,4,1,5,0,10]) #respuesta al escalón del slit
h = np.diff(he)#respuesta al impulso del slit
print(h)
```

La salida del sistema se puede obtener entonces mediante la operación de convolución:

$y[n]=x[n]*h[n]=\sum_{k=-\infty}^{\infty}x[k]h[n-k]$

```python
y = np.convolve(xq,h,mode='same')#respuesta del slit -> se asegura tamaños iguales con mode=same
#graficar señal
plt.title('Salida del sistema')
plt.stem(t,y,'b',label='$y[n]$')
plt.legend()
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.show()
```
